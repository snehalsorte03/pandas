#### Conditional Changes

```python
import pandas as pd
```

```python
details= {
    'Name': ['Ankit','Aishhwarya','Ankit','Aishwarya','Shaurya','Shivangi'],
    'Maths': [80,40,50,70,42,61],
    'Sci' : [83,21,43,61,72,71],
    'Hindi' : [53,21,73,11,92,11],
    'Chem': [73,41,43,41,82,91],
    'Bio' : [93,41,23,41,45,51],
    'Sanskrit': [73,81,53,41,72,21]
}
d1=pd.DataFrame(details)
d1
```

```python
d1["total"]=0
```

```python
d1
```

```python
d1["total"]=d1["Maths"]+d1["Sci"]+d1["Hindi"]+d1["Chem"]+d1["Bio"]+d1["Sanskrit"]
```

```python
d1
```

```python
d1["grade"]=0
```

```python
d1
```

```python
d1["%"]=(d1["total"]/600)*100
```

```python
d1
```

- less than 45-Fail
- 45-60 pass
- 60-70 first class
- 70- distinction


```python
d1.loc[d1["%"]<45]
d1.loc[d1["%"]<45,"grade"]="FAIL"
```

```python
d1
```

```python
d1.loc[(d1["%"]>45) & (d1["%"]<60)]
d1.loc[(d1["%"]>45) & (d1["%"]<60),"grade"]="PASS"
```

```python
(d1["%"]>45) & (d1["%"]<60)
```

```python
d1
```

```python
d1.loc[(d1["%"]>70)]

d1.loc[(d1["%"]>70),"grade"]="DISTINCION"

d1.loc[(d1["%"]>60) & (d1["%"]<70)]

d1.loc[(d1["%"]>60) & (d1["%"]<70),"grade"]="FIRST CLASS"
```
