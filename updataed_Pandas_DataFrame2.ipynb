### Merge

- outer join- use union of keys
- inner join- intersection of keys
- left join- use keys from left table/dataframe/object
- right join-use keys from right object


```python
import pandas as pd
df1=pd.DataFrame({
    "city":["new york","chicago"],
    "temperature":[21,14],
})
df1
```

```python
df2 =pd.DataFrame({
    "city":["new york","orlando","chicago"],
    "humidity":[68,75,23]
})
df2
```

```python
df3=pd.merge(df1,df2, on="city",how="inner")
df3
```

```python
df3=pd.merge(df1,df2, on="city",how="outer",indicator=True)  #its indicates the type of merging.
df3
```

### Suffix

```python
df1=pd.DataFrame({
    "city":["new york","chicago","orlando","baltimore"],
    "temperature":[21,14,35,38],
    "humidity":[65,68,71,75]
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicango","new york","san diego"],
    "temperature":[21,14,35],
    "humidity":[65,68,71]
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",suffixes=('first','second'))
df3
```

- Both join and merge can be used to combines two dataframes but the join method combines two dataframes on the basis of their indexes whereas the merge method is more versatile and allows us to specify columns beside the index to join on for both dataframes.

### join

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({
    "city":["new york","chicago","orlando","Dallas"],
    "temperature": [21,14,35,23]
})
df1
```

```python
df1.set_index('city',inplace=True)
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","Los Angeles"],
    "humidity":[65,68,23]
})
df2
```

```python
df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2,how="left") #takes all data from dataframe1 and common from dataframe 2
```

```python
df1.join(df2,how="right")  #takes all data from dataframe2 and common from dataframe 1
```

```python
df1.join(df2,how="inner")
```

```python
df1.join(df2,how="outer") #full join
```
